```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

# === Chargement des donn√©es ===
df = pd.read_csv("../data/cleaned/donnees_climat_evaporation.csv")
print(f"Donn√©es charg√©es : {df.shape[0]} lignes, {df.shape[1]} colonnes")
```

```python
print("\nüîç Aper√ßu des 5 premi√®res lignes :")
display(df.head())

print("\n‚ÑπÔ∏è Informations g√©n√©rales sur les colonnes :")
df.info()

print("\nüìä Statistiques descriptives (variables num√©riques) :")
display(df.describe())
```

```python
print("üìâ Distribution des variables num√©riques :")
df.select_dtypes(include='number').hist(bins=30, figsize=(14,10), color="#1f77b4", edgecolor='black')
plt.suptitle("Distribution des variables", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
print("üîó Calcul de la matrice de corr√©lation...")

corr = df.select_dtypes(include='number').corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("Matrice de corr√©lation entre les variables num√©riques", fontsize=16)
plt.show()
```

```python
print("üìå Corr√©lations les plus fortes avec 'Evap_estimee' :")
display(corr['Evap_estimee'].sort_values(ascending=False))
```

```python
# Choisir quelques variables explicatives pertinentes
vars_explicatives = ['Insolation', 'TCM=(Tmin+Tmax)/2', 'Vitesse_moy', 'Humidit√© relative MIN']

print("üìà Graphiques de dispersion pour analyser visuellement les relations avec 'Evap_estimee'")

for var in vars_explicatives:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=var, y='Evap_estimee', data=df, alpha=0.5)
    plt.title(f"Evap_estimee en fonction de {var}")
    plt.xlabel(var)
    plt.ylabel("Evap_estimee (mm/jour)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
